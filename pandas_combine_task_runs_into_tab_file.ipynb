```python
import pandas as pd
import numpy as np
import os
import glob
```

```python
basepath='/Users/jennygilbert/Documents/RW_bevel/task_logs'
list = glob.glob('task_logs/*.txt')
#print(filenames)

list_of_dfs=[pd.read_csv(file, delimiter = '\t', header=None, names=('subjID','type','choice','reward')) for file in list]
list_of_dfs
```

```python
filename= "all_clean.txt"
combined_df = pd.concat(list_of_dfs, ignore_index=True)
combined_df.to_csv(filename, sep="\t", index=False, header=('subjID','type','choice','reward'))
```

```python
dict = {}
for filename in list:
    id = filename.split("/")[1].split("_")[0]
    if id not in dict:
        dict[id] = []
        #print(id)
    else:
        pass
    dict[id].append(filename)
```

```python
for subj in dict:
    list_dfs = [pd.read_csv(file, delimiter = '\t', header=None) for file in dict[subj]]
    #print(list_dfs)
    combined_df = pd.concat(list_dfs, ignore_index=True)
    #print(combined_df)
    filename= subj+"_clean.txt"
    #print(filename)
    combined_df.to_csv(filename, sep="\t", index=False, header=('subjID','type','choice','reward'))
```
